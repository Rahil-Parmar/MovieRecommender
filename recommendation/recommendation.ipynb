```python
drop_column = ['IMDb_Id','Crew','Collection','Release_Date','Release_Status','Original_Language','Languages_Spoken','Runtime','Popularity','Rating_average','Rating_Count','Production_Companies','Country_of_Origin','Budget','Revenue']
data.drop(drop_column,axis = 1,inplace=True)
```

```python
data = data.fillna('')
```

```python
def s2w(sentence):
    return [word.strip() for word in sentence.split("|")]
```

```python
data['Genres'] = data['Genres'].apply(lambda x:s2w(x))
data['Cast'] = data['Cast'].apply(lambda x:s2w(x))
```

```python
data.head()
```

```python
data['Cast'] = data['Cast'].apply(lambda x: x[:5])
```

```python
data['cast'] = data['Cast']
```

```python
swords = nltk.corpus.stopwords.words('english')
data['Overview'] = data['Overview'].apply(lambda x: [w for w in x.split() if w not in swords])
data['Tagline'] = data['Tagline'].apply(lambda x: [w for w in x.split() if w not in swords])
```

```python
data['Genres'] = data['Genres'].apply(lambda x: [i.replace(' ','') for i in x])
data['Cast'] = data['Cast'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
data['tags'] = data['Overview'] + data['Genres'] + data['Cast'] + data['Tagline']
```

```python
data.head()
```

```python
new_data = data[['TMDb_Id','Title','tags','cast']]
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x: [i.replace(',','') for i in x])
```

```python
new_data.head()
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x: " ".join(x))
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  l = []
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```

```python
new_data['tags'] = new_data['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vectors)
```
